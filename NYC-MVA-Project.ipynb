```python
# drop rows with null contributing factor values

contributing_factor_df = location_df[location_df.contributing_factor_vehicle_1.notnull()]

contributing_factor_df
```

```python
# drop rows with null on street name values

on_street_df = contributing_factor_df[contributing_factor_df.on_street_name.notnull()]

on_street_df
```

```python
# drop rows with null off street name values

off_street_df = on_street_df[on_street_df.off_street_name.notnull()]

off_street_df
```

```python
test_df = off_street_df.drop(off_street_df[(off_street_df['contributing_factor_vehicle_1'] == 'Unspecified')].index)
test_df
```

```python
# Format data to remove timestamp

df2 = off_street_df.copy()

df2['crash_date'] = pd.to_datetime(df2['crash_date']).dt.normalize()
df2
```
